# Techniques outside of class:

## Maximum Drawdown
Maximum drawdown is a metric used to assess an investment's risk, measuring the greatest peak-to-trough loss in an investment's value over a given time frame. It is often used in conjunction with other risk metrics, such as volatility, and can be used to contrast the risk of various investment options.
## Rolling Statistics
Rolling statistics are calculations on a sliding window with a fixed size over a time period or other sequential data. They can be used to spot patterns, trends, and alterations in the data over time. Examples of rolling statistics include rolling mean, rolling standard deviation, rolling minimum, rolling maximum, and rolling correlation. Pandas and NumPy are two libraries available in Python that include routines for quickly calculating rolling statistics.

```python

# Calculate the cumulative returns
ETF_df['cumulative_returns'] = (1 + ETF_df['Vanguard S&P 500 Price (Rand)']).cumprod()

# Calculate the running maximum of cumulative returns
ETF_df['running_maximum'] = ETF_df['cumulative_returns'].cummax()

# Calculate the drawdown as the percentage decline from the running maximum
ETF_df['drawdown'] = (ETF_df['running_maximum'] - ETF_df['cumulative_returns']) / ETF_df['running_maximum']

# Calculate the maximum drawdown
max_drawdown = np.max(ETF_df['drawdown'])

# Print the maximum drawdown
print("Maximum Drawdown:", max_drawdown)
```

```python

fig, ax = plt.subplots()

# Plot the cumulative returns and running maximum on the first axis
ax.plot(ETF_df['cumulative_returns'], label='Cumulative Returns')
ax.plot(ETF_df['running_maximum'], label='Running Maximum')

ax.set_title('Maximum Drawdown')
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Calculate and plot the drawdowns as a separate subplot
drawdown = (ETF_df['running_maximum'] - ETF_df['cumulative_returns']) / ETF_df['running_maximum']
ax2 = ax.twinx()
ax2.fill_between(drawdown.index, drawdown, alpha=.2, color='red', label='Drawdown')
ax2.set_ylim([0,1])
ax2.set_ylabel('Drawdown')

ax.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```
